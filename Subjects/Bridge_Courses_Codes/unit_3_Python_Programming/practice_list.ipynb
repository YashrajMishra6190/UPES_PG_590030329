# List
---
- List are mutable(changeable).
- Empty List: list = []
- Allows Duplicate values
- ordered collection
- List allows different data types

## Add new Items in list 
- .append(): add items at last
- .insert(index, "values"): add elements when index given

## Modify and Remove Items: 
- .remove(): remove the last element of list 
- .pop(index): remove the element mentioned as index value

## Merge List

- list1.extend(list2)
- list1 + list2: "+" concatenation

## Count
- .count("sample")

## Indexing 
- .index("sample")

## Sorting
- .sort() # by default ascending 
- .sort(reverse = True) # for descending terms 

## Copy Function 
- .copy() # create new list 

## Clear() Function 
- .clear()

## List Comprehensions

- list = [execcute for i in range(5)]

# Pratcice List: 

- Max 
- Minimum 
- sorting 

```python
# find the maximum of list 

list_max = [1,4,25,52,23,53,18,324]
max_num = list_max[0]
for numbers in list_max:
    if numbers > max_num:
        max_num = numbers
print(max_num)

```
