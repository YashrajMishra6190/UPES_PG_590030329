# SETS

----
- Unordered 
- mutable
- no indexing 
- not allow duplicate values



| Method                   | Purpose                                     |
| ------------------------ | ------------------------------------------- |
| `set()`                  | Create an empty set                         |
| `{1,2,3}`                | Create a set with elements                  |
| `add(x)`                 | Add one element                             |
| `update(iterable)`       | Add multiple elements                       |
| `remove(x)`              | Remove an element (raises error if missing) |
| `discard(x)`             | Remove an element (no error if missing)     |
| `pop()`                  | Remove and return a random element          |
| `clear()`                | Remove all elements                         |
| `len(set)`               | Get number of elements                      |
| `copy()`                 | Copy a set                                  |
| `union()`                | Combine sets                                |
| `intersection()`         | Common elements                             |
| `difference()`           | Elements only in first set                  |
| `symmetric_difference()` | Elements in either set but not both         |
| `issubset()`             | Check subset                                |
| `issuperset()`           | Check superset                              |
| `isdisjoint()`           | Check if no common elements                 |


```python
# Empty set 

empty_set = set()
# avoid using set = {} -- As it is for dictionary empty value
```

```python
#addition of set()
set_1 = {1,2,34,4}
set_1.add(21)
```

```python
set_1
```

```python
# update([list])

set_1.update([23,11,99])
set_1
```

```python
# Remove 
set_1.remove(99)
print(set_1)
# Discard
set_1.discard(11)
set_1
```

```python
len(set_1)
```

```python
# Intersection returns common values
set1 = {1,2,3,7,4,9}
set2 = {1,24,2,4,9}
```

```python
set1.intersection(set2)
```

```python
# Difference
set1.difference(set2)
```

```python
# Symmetric Difference 
set1.symmetric_difference(set2)
```

```python
# Union 
set3 = set1.union(set2)
set3
```

```python
# subset 
set1.issubset(set3)
 
```

```python
# Superset 
set3.issuperset(set2)
```

```python
# isdisjoint

set1.isdisjoint(set2)
```

```python
set_2 = {1,2,3}

set_3 = {4,5,6}

set_2.isdisjoint(set_3)
```

```python
set_2.union(set_3)
```
