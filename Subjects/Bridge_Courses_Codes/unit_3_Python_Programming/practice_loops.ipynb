Workout Timer Countdown
In this example, You are building a simple workout timer that counts down from a user-defined time. The timer will display the remaining seconds after each iteration until it reaches zero.

Task

Takes user input for the workout duration in seconds, stores it in a variable called workout_time, and converts it to an integer.
Uses a while loop to count down from workout_time to zero.
Prints the remaining time in each iteration using f-string in the format: Time remaining: {workout_time} seconds
After the countdown reaches zero, print the final message: Workout complete! Great job!

Sample 1:
Input

Output
2

Time remaining: 2 seconds
Time remaining: 1 seconds
Time remaining: 0 seconds
Workout complete! Great job!

```python
workout_time = int(input().strip())
while workout_time >=0:
    print(f"Time remaining: {workout_time} seconds")
    workout_time -=1
print("Workout complete! Grteat Job!")
```

Ticket Counter System
In this task, A ticket counter allows customers to book tickets sequentially from ticket number 1 up to a user-defined limit. The system ensures that all tickets are issued in order, and once the last ticket is issued, it prints a confirmation message.

Task

Take user input for total_tickets, convert it to an integer.
Initialize ticket_number = 1 to start counting from the first ticket.
Use a while loop to issue tickets from 1 to total_tickets (using <= operator).
Print each issued ticket using an f-string: Ticket {ticket_number} issued.
Increment ticket_number by 1 in each iteration to move to the next ticket.
Once all tickets are issued, the program prints: All tickets issued. Counter closed.


```python
ticket_counter = int(input().strip())
i = 1
while i <= ticket_counter:
    print(f"Ticket {i} issued.")
    i +=1
print("All tickets issued. Counter closed.")
```

Student Attendance Tracker
A teacher tracks attendance for user-defined days, skipping Day 3 (holiday).

Write a Python program to take the student's name and total days as input and print attendance, skipping Day 3 using a while loop and continue.

Task

Take user input for name using input().strip().
Take user input for total_days using input(), convert it to an integer.
Initialize day to 1 and use a while loop to iterate from 1 to total_days.
If day == 3, increment it and use continue to skip printing.
Print attendance using an f-string: Attendance recorded for {name} on Day {day}.
Increment day after each iteration.
Ensure the loop stops after reaching total_days, skipping only Day 3.


```python
student_name = input() 
total_days = int(input().strip())

day = 1
while day <= total_days:
    if day == 3:
        day +=1
        continue
    print(f"Attendance recorded for {student_name} on Day {day}")
    day +=1
    
```

Multiplication Table with Even-Odd Condition
In this task, you will generate multiplication tables for numbers within a specified range.
For each user input 
n - you need to output multiplication tables upto n.

Each multiplication result will be marked as either "E" (even) or "O" (odd) based on its parity.


for n  = 3

1(O)  2(E)  3(O) 
2(E)  4(E)  6(E) 
3(O)  6(E)  9(O)

```python
n = 3 
limit = n*n # 9

for i in range(1, limit+1):
    j = n*i
    if j%2 ==0:
        print(f"{i}(E)", end = " ")
    else: print(f"{i}(O)", end = " ")
print()
```

Calculating Factorials for Series Summation
In this task, we will compute the sum of the factorial series using a for loop in Python.
The series is defined as:
S = 0! + 1! +2! +3! + ⋯ + n!
where n! (n factorial) is the product of all positive integers from 1 to n.



```python
n = 4
total_sum = 0  # Avoid using 'sum' as a variable name since it's a built-in Python function

# Loop from 1 to n (1, 2, 3, 4)
for i in range(0, n + 1):
    fact = 1
    # Calculate the factorial for the current number 'i'
    for j in range(1, i + 1):
        fact *= j
    # Add the current factorial to our total sum
    total_sum += fact

print(total_sum)
```
