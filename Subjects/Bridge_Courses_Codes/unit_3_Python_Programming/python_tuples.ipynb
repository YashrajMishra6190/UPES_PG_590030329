```python
tup_1[0]
```

```python
# Without inbuilt fucntion 

maximum = tup_1[0]
for i in tup_1:
    if i > maximum:
        maximum = i
print(maximum)
```

```python
# Inbuilt function to find min 

min(tup_1)
```

```python
# Without inbuilt fucntion 
minimum = tup_1[0]
for i in tup_1:
    if i < minimum:
        minimum = i
print(minimum)
```
