```python
nums = [19, 2, 4, 23]

max_value = nums[0]
max_index = 0

for i in range(len(nums)):
    if nums[i] > max_value:
        max_value = nums[i]
        max_index = i

print("Max value:", max_value)
print("Index:", max_index)
```

```python
chars = ['a', 'v', 'c']

n = len(chars)

for i in range(n):
    for j in range(0, n - i - 1):
        if ord(chars[j]) > ord(chars[j + 1]):
            chars[j], chars[j + 1] = chars[j + 1], chars[j]

print(chars)
```

```python
names = ["abcd", "trees", "freak"]

# Step 1: Extract the first characters using a for loop
first_chars = []
for name in names:
    first_chars.append(name[0])

# Step 2: Sort the list using a for loop (Bubble Sort algorithm)
n = len(first_chars)
for i in range(n):
    for j in range(0, n - i - 1):
        # Compare adjacent elements
        if first_chars[j] > first_chars[j + 1]:
            # Swap them if they are in the wrong order
            first_chars[j], first_chars[j + 1] = first_chars[j + 1], first_chars[j]

print(first_chars)
```
